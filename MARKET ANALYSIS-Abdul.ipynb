##  4.Cleaning. Dealing with duplicates, null values and outliers

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
#Transforming " Income " to "Income"
df.rename(columns = {" Income ":"Income"},inplace = True)
```

```python
df["Income"] = df["Income"].str.replace("$","")
df["Income"] = df["Income"].str.replace(",","").astype(float)
```

#### Checking Income's distribution 

```python
sns.boxplot(x = df["Income"]);
```

```python
fig = plt.figure(figsize = (10,5));
sns.histplot(x = df["Income"]);
```

##### Will fill income's null values with median to avoid effects of outliers 

```python
n = df["Income"].median()
```

```python
df["Income"].fillna(n, inplace = True)
```

##### Removing outliers

```python
df.shape
```

```python
df.head(1)
```

```python
df.plot(subplots = True, layout = (6,5), kind = "box" ,figsize=(12,14),patch_artist=True);
plt.subplots_adjust(wspace=0.5);
```

```python
ab = (df["Income"]<600000) & (df["Year_Birth"]>1910)
```

```python
df = df[ab]
```

###### Your data is now clean

# 4.Feature Engineering

###### Creating new features

```python
df.head(4)
```

```python
df["Kid_Teen_home"] = df["Kidhome"] + df["Teenhome"]
```

```python
df['Total_Spent']=(df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds'])
df['Total_Purchases']=(df['NumDealsPurchases']+df['NumWebPurchases']+df['NumCatalogPurchases']+df['NumStorePurchases']+df['NumWebVisitsMonth'])
df["TotalCampaignsAcc"] =df["AcceptedCmp1"]+df["AcceptedCmp2"]+df["AcceptedCmp3"]+df["AcceptedCmp4"]+df["AcceptedCmp5"]+df["Response"]
```

##### Changing Dt_Customer to datetime

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
```

##### We can now get the age of customers

```python
df["Age"]= df["Dt_Customer"].dt.year - df["Year_Birth"]
```

### Drop the unwanted features

```python
df.head()
```

```python
df.drop(columns = ["ID","Year_Birth","Kidhome","Teenhome","Dt_Customer"],inplace = True)
```

```python
#Expoted the clean for further analysis in tableau
df.to_csv("clean_data")
```
